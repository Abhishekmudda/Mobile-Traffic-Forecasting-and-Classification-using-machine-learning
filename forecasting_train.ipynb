```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.models import load_model
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model

file_path=r"C:\Users\abhim\OneDrive\Desktop\forecasting.csv"
data = pd.read_csv(file_path)

data['Time'] = pd.to_datetime(data['Time'])

data.set_index('Time', inplace=True)

print(data.isnull().sum())

data.fillna(method='ffill', inplace=True)

# Print the updated dataframe
print(data.head())

# Assuming 'Length' is the column to predict
traffic_data = data[['Length']]

# Resample data to 1-minute frequency
traffic_data = traffic_data.resample('1min').mean()
traffic_data = traffic_data.fillna(traffic_data.mean())

# Normalize the data
scaler = StandardScaler()
traffic_data_scaled = scaler.fit_transform(traffic_data)

# Split the data into train and test
train_size = int(len(traffic_data_scaled) * 0.8)
train, test = traffic_data_scaled[:train_size], traffic_data_scaled[train_size:]

# Reshape data for LSTM
train = np.reshape(train, (train.shape[0], 1, train.shape[1]))
test = np.reshape(test, (test.shape[0], 1, test.shape[1]))

# Build the LSTM model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(train.shape[1], train.shape[2])))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')

# Train the model
model.fit(train, train, epochs=10, batch_size=1, verbose=2)

# Make predictions
train_predict = model.predict(train)
test_predict = model.predict(test)

# Invert scaling
train_predict = scaler.inverse_transform(train_predict)
test_predict = scaler.inverse_transform(test_predict)

# Calculate RMSE
train_score = np.sqrt(mean_squared_error(train[:, 0, 0], train_predict[:, 0]))
test_score = np.sqrt(mean_squared_error(test[:, 0, 0], test_predict[:, 0]))
print('Train Score: %.2f RMSE' % (train_score))
print('Test Score: %.2f RMSE' % (test_score))

# Plot the results
plt.figure(figsize=(16, 8))

# Plot actual values
plt.plot(traffic_data.index, scaler.inverse_transform(traffic_data_scaled), label='Actual', color='blue')

# Plot predicted values for the training set
plt.plot(traffic_data.index[:train_size], train_predict, label='Train Predict', color='green')

# Plot predicted values for the testing set
plt.plot(traffic_data.index[train_size:], test_predict, label='Test Predict', color='red')

plt.legend()
plt.xlabel('Time')
plt.ylabel('Traffic Volume')
plt.title('Actual vs Predicted Traffic Volume')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

plt.figure(figsize=(16, 8))

# Plot original training values
plt.subplot(2, 1, 1)
plt.plot(traffic_data.index[:train_size], scaler.inverse_transform(traffic_data_scaled)[:train_size], label='Actual Train', color='blue')
plt.xlabel('Time')
plt.ylabel('Traffic Volume')
plt.title('Actual Training Traffic Volume (1-min intervals)')
plt.xticks(rotation=90)
plt.legend()

# Plot predicted training values
plt.subplot(2, 1, 2)
plt.plot(traffic_data.index[:train_size], train_predict, label='Train Predict', color='green')
plt.xlabel('Time')
plt.ylabel('Traffic Volume')
plt.title('Predicted Training Traffic Volume (1-min intervals)')
plt.xticks(rotation=90)
plt.legend()

plt.tight_layout()
plt.show()

# Plot for testing set
plt.figure(figsize=(16, 8))

# Plot original testing values
plt.subplot(2, 1, 1)
plt.plot(traffic_data.index[train_size:], scaler.inverse_transform(traffic_data_scaled)[train_size:], label='Actual Test', color='blue')
plt.xlabel('Time')
plt.ylabel('Traffic Volume')
plt.title('Actual Testing Traffic Volume (1-min intervals)')
plt.xticks(rotation=90)
plt.legend()

# Plot predicted testing values
plt.subplot(2, 1, 2)
plt.plot(traffic_data.index[train_size:], test_predict, label='Test Predict', color='red')
plt.xlabel('Time')
plt.ylabel('Traffic Volume')
plt.title('Predicted Testing Traffic Volume (1-min intervals)')
plt.xticks(rotation=90)
plt.legend()

plt.tight_layout()
plt.show()
```
